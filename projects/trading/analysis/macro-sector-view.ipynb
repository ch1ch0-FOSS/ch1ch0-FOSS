```python
import subprocess
import json

# Test with longer timeout (Plutus cold-start can take 90+ seconds)
prompt = "What is an accumulation pattern in trading?"

print("Starting Plutus... (this may take 60-90 seconds on first run)")

try:
    result = subprocess.run(
        ['ollama', 'run', '0xroyce/plutus'],
        input=prompt.encode(),
        capture_output=True,
        timeout=120  # Increased from 60 to 120 seconds
    )
    
    response = result.stdout.decode('utf-8', errors='ignore')
    print("\n✓ Plutus Response:")
    print(response)
    print("\n✓ LLM Integration Working!")
    
except subprocess.TimeoutExpired:
    print("✗ Timeout - Plutus is still loading or system is memory-constrained")
    print("Try running from terminal instead:")
    print("  ollama run 0xroyce/plutus")
    
except Exception as e:
    print(f"✗ Error: {str(e)}")
```
