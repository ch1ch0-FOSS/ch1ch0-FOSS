```python
# CELL 1
import pandas as pd
import subprocess
from datetime import datetime

# QQQ top holdings (you can add more ETFs)
holdings_map = {
    'QQQ': ['MSFT', 'AAPL', 'NVDA', 'GOOGL', 'META', 'TSLA', 'AMZN', 'AVGO'],
    'SPY': ['AAPL', 'MSFT', 'NVDA', 'GOOGL', 'AMZN', 'TSLA', 'META', 'BERKB'],
    'XLK': ['MSFT', 'AAPL', 'NVDA', 'GOOGL', 'META', 'ADBE', 'CRM', 'CSCO'],
    'XLF': ['JPM', 'BAC', 'WFC', 'GS', 'BLK', 'BK', 'MS', 'TROW'],
}

# Signal ETFs from accumulation tracker
signal_etfs = ['QQQ', 'SPY', 'XLK', 'XLF']

print("="*70)
print("HOLDINGS DRILL - Finding Individual Stock Opportunities")
print("="*70)

for etf in signal_etfs:
    holdings = holdings_map.get(etf, [])
    print(f"\n{etf} Top Holdings:")
    for i, holding in enumerate(holdings, 1):
        print(f"  {i}. {holding}")
```

```python
# CELL 2
print("\n" + "="*70)
print("PLUTUS: Holdings Analysis")
print("="*70)

# Generate Plutus recommendations for holdings
for etf in signal_etfs:
    holdings = holdings_map.get(etf, [])
    holdings_str = ", ".join(holdings[:5])  # Top 5
    
    prompt = f"""The ETF {etf} is showing institutional accumulation signal (2.1-2.8x volume).

Top holdings are: {holdings_str}

Which of these holdings would be best bull call spread candidates?
Rate each by:
1. Likely to be the driver of the ETF move
2. Options liquidity (can we trade 0DTE/1DTE?)
3. Risk/reward potential

Recommend top 2 holdings to trade instead of the ETF."""

    print(f"\n{etf} Analysis:")
    print("-" * 70)
    
    try:
        result = subprocess.run(
            ['ollama', 'run', '0xroyce/plutus'],
            input=prompt.encode(),
            capture_output=True,
            timeout=90
        )
        
        analysis = result.stdout.decode('utf-8', errors='ignore')
        print(analysis)
        
    except Exception as e:
        print(f"Error: {str(e)}")
```

```python
# CELL 3
print("\n" + "="*70)
print("NEXT STEPS")
print("="*70)

workflow = """
Based on Plutus recommendations:

1. Identify top 2 holdings per ETF signal
2. Search each on TradingView
3. Check charts + options chain liquidity
4. Drill down to individual stock bull calls
5. Execute on highest conviction holdings

Example flow:
- QQQ signal (2.1x) → Plutus recommends MSFT + NVDA
- Check MSFT daily chart → Find entry
- Check MSFT options chain → Confirm liquidity for 0DTE
- Execute MSFT bull call instead of QQQ ETF (tighter focus)
"""

print(workflow)
```
